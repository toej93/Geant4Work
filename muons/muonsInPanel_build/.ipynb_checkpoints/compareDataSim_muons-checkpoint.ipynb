```python
folder = "/home/jat89/Software/Geant4Work/muons/muonsInPanel_build/"
file = ROOT.TFile(folder + "muons_horizontal.root", "READ")
file2 = ROOT.TFile(folder + "muon_vertical.root", "READ")

file.ls()
```

```python
# gROOT.SetStyle("ATLAS");
h1 = file.Get("10")
c2 = ROOT.TCanvas("c2", "c2", 700, 500)
h1.Draw("")
c2.Draw()
c2.Delete
```

```python
a = file.Get("ntuple")
a2 = file2.Get("ntuple")

a.ls()
```

```python
tree = a.Get("parTup")
tree2 = a2.Get("parTup")
```

```python
energy = []
for event in tree:
    energy.append(event.Edeposited)
    
energy_vertical = []
for event in tree2:
    energy_vertical.append(event.Edeposited)
```

### Open files of data that Samantha took

```python
data_vert = np.load('VerticalFilter.npy')
data_horz = np.load('HorizontalFilter.npy')
```

```python
plt.figure(figsize = (8,5))
plt.hist(energy, bins = 100, range = (0,80), histtype="step", lw = 1.7, label = "Horizontal panel", density = True);
plt.hist(energy_vertical, bins = 100, range = (0,80), histtype="step", lw = 1.7, label = "Vertical panel", density = True);
# plt.hist(data_vert, bins = 100, range = (0,40), histtype="step", lw = 1.7, label = "Vertical panel [data]", density = True);

plt.yscale("log")
plt.xlabel("Deposited energy [MeV]")
plt.title("Flux of muons on panel (simulation)")
plt.grid(alpha = 0.3)
plt.legend(fontsize = 14)
# plt.axvline(5, c = "red")
plt.tight_layout()
plt.savefig("muonsPanel_verHorComparison.pdf")
```

```python
plt.figure(figsize = (7,5))
plt.hist(energy, bins = 100, range = (4,30), histtype="step", lw = 1.7, label = "Simulation", density = True);
plt.hist(data_horz, bins = 100, range = (4,30), histtype="step", lw = 1.7, label = "Data", density = True);
# plt.hist(data_vert, bins = 100, range = (0,40), histtype="step", lw = 1.7, label = "Vertical panel [data]", density = True);

plt.yscale("log")
plt.xlabel("Deposited energy [MeV]")
plt.title("Muons on panel (horizontal configuration)")
plt.grid(alpha = 0.3)
plt.legend(fontsize = 14)
# plt.axvline(5, c = "red")
plt.tight_layout()
plt.savefig("muonsPanel_HorDataSim.pdf")
```
